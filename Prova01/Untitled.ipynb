```python
import matplotlib.pyplot as plt
import numpy as np
from scipy.fft import fft
from scipy.signal import stft
```

```python
def graph(f0, ft, T):
    c = (ft - f0) / T
        
    n_samples = 10000
    x = np.linspace(0, 1, n_samples)
    t = np.linspace(0, T, n_samples)
    
    f = (c*t) + f0
    plt.plot(f)

    curve = np.sin(2 * np.pi * x * f)

    plt.figure(figsize=[15, 5])
    plt.plot(t, curve,  label="Curve")
    plt.legend()
    plt.show()
 
    plt.figure(figsize=[15, 5])
    plt.plot(np.abs(fft(curve)), label="FFT")
    plt.legend()
    plt.show()
    
    f, t, Zxx = stft(curve)

    plt.figure(figsize=(20, 10))

    plt.pcolormesh(t, f, np.log(np.abs(Zxx)), shading='gouraud')
    plt.title('STFT Magnitude')
    plt.ylabel('Frequency [Hz]')
    plt.xlabel('Time [sec]')
    plt.show()
    
graph(10, 100, 100)
    
```
